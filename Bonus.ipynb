# Problem 1
Compute svd of the following matrix:
$$ 
A = \begin{pmatrix}
1 & 2 & 3 & \ldots & n \\
1 & 2 & 3 & \ldots & n
\end{pmatrix}
$$
# Solution
$$
A = \begin{pmatrix} 1 \\
1 \end{pmatrix} \begin{pmatrix}
1 & 2 & 3 & \ldots & n
\end{pmatrix}
$$
$$
A^TA = \begin{pmatrix}
1 \\
2 \\
3 \\
\ldots \\
n
\end{pmatrix} \begin{pmatrix} 1 & 1 \end{pmatrix} 
\begin{pmatrix} 1 \\
1 \end{pmatrix} \begin{pmatrix}
1 & 2 & 3 & \ldots & n
\end{pmatrix} = 2\begin{pmatrix}
1 \\
2 \\
3 \\
\ldots \\
n
\end{pmatrix}
 \begin{pmatrix}
1 & 2 & 3 & \ldots & n
\end{pmatrix} = 2vv^T
$$
$$
v = \begin{pmatrix}
1 \\
2 \\
3 \\
\ldots \\
n
\end{pmatrix}$$
It is easy to see that $2vv^T$ as a matrix has $n - 1$ zero eigenvalues ($n - 1$ vectors orthogonal to $v$) and 1 nonzero eigenvalue for $v$:
$$2vv^Tv = 2|v|^2 v$$
$$\sigma_1 = \sqrt{2}|v| = \sqrt{2} \sqrt{\sum_{i=1}^n i^2} = \sqrt{\frac{n(n+1)(2n+1)}{3}}$$
$$\Sigma = \begin{pmatrix}
\sigma_1
\end{pmatrix}$$
$$V^T = \frac{\sqrt{2}}{\sigma_1} \begin{pmatrix}
1 & 2 & \ldots & n
\end{pmatrix}$$
$$u = \frac{1}{\sigma_1} Av = \frac{1}{\sqrt{2}} \begin{pmatrix}
1 \\
1
\end{pmatrix}$$
$$A = \begin{pmatrix}
\frac{1}{\sqrt{2}} \\
\frac{1}{\sqrt{2}}
\end{pmatrix} \begin{pmatrix}
\sigma_1
\end{pmatrix} \begin{pmatrix}
\frac{\sqrt{2}}{\sigma_1} & \frac{2\sqrt{2}}{\sigma_1} & \ldots & \frac{n\sqrt{2}}{\sigma_1}
\end{pmatrix}$$

# Problem 2
Suppose you have an unknown quadratic form (symmetric matrix $A$) that acts as a black box, i.e. you can only input any vector $v$ and get the result $v^TAv$, but you don't know how $A$ looks. What is the minimum number of vectors you should pass to the black box to determine whether $A$ is positive definite or not?
# Solution
By passing $\frac{n(n + 1)}{2}$ vectors you can recover all the elements of $A$ and determine if it is positive definite or not. (First recover diagonal elements by passing the basis vectors, then recover everything else). <br>
Let's prove that we can't do any better. Suppose we passed less than $\frac{n(n + 1)}{2}$ vectors, then some of the elements of $A$ are unknown to us. <br>
If we don't know some diagonal element $a_{ii}$ then it can be negative and e_i^TAe_i = a_ii < 0, thus matrix won't be positive definite. <br>
If we don't know some off-diagonal element $a_{ij}$ then $(e_i + e_j)^TA(e_i + e_j) = a_{ii} + a_{jj} + 2a_{ij}$ can be negative and matrix won't be positive definite.

# Problem 3
Let $A$ be invertible matrix with all positive entries ($a_{ij} > 0$). Prove that $A^{-1}$ has no more than $n^2 - 2n$ zeros.
# Solution
$$
A = \begin{pmatrix}
a_{11} & \ldots & a_{1n} \\
& \ldots & \\
a_{n1} & \ldots & a_{nn}
\end{pmatrix} \quad A^{-1} = \begin{pmatrix}
a'_{11} & \ldots & a'_{1n} \\
& \ldots & \\
a'_{n1} & \ldots & a'_{nn}
\end{pmatrix}
$$
$$
AA^{-1} = E
$$
$$
\sum_{j}a_{ij}a'_{jk} = \delta_{ik}
$$
Suppose that $A^{-1}$ has at least $n^2 - 2n + 1$ zeros, it means that there is a column with only 1 nonzero element (we can't have all zeros columns due to nondegeneracy). Suppose it is the $l$-th column.
$$
\sum_{j}a_{ij}a'_{jl} = a_{ik}a'_{kl} = \delta_{ik}
$$
Take $i \neq l$
$$ a_{ik}a'_{kl} = 0 $$
On the other hand $a'_{kl} \neq 0$ and $a_{ik} > 0$. Contradiction.

# Problem 4
Let A be a real square matrix. Prove that if $A^T = p(A)$ for some polynomial $p(x): p(0) \neq 0$ then $A$ is invertible.
# Solution
$$ p(x) = \sum_{i=1}^m a_i x^i $$
Suppose that $A$ is degenerate $\Leftrightarrow$ $\exists v: Av = 0, \|v|\ \neq 0$ <br>
Consider $v^TA^Tv = v^Tp(A)v$
$$v^TA^Tv = (Av)^Tv = 0$$
$$v^Tp(A)v = v^T\left(\sum_{i=1}^m a_i A^iv \right) = a_0v^Tv = a_0\|v\|^2 \neq 0$$
Contradiction.
